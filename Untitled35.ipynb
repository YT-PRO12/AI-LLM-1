```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load data
df = pd.read_csv(r"C:\Users\yatharth\Downloads\Zomato-data-.csv")

# Remove duplicates
df = df.drop_duplicates()

# CLEAN RATE COLUMN PROPERLY 
df['rate'] = df['rate'].astype(str)

# Remove '/5' and unwanted text
df['rate'] = df['rate'].str.replace('/5', '')
df['rate'] = df['rate'].replace(['NEW', '-', 'nan'], None)

df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

#  CLEAN COST COLUMN 
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(str)
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',', '')
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'], errors='coerce')

#  CLEAN ONLINE ORDER 
df['online_order'] = df['online_order'].map({'Yes':1, 'No':0})

#  KEEP ONLY IMPORTANT COLUMNS 
df = df[['approx_cost(for two people)', 'online_order', 'rate']]

# DROP ONLY ROWS WHERE TARGET IS MISSING
df = df[df['rate'].notna()]

# FILL REMAINING MISSING VALUES 
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].fillna(df['approx_cost(for two people)'].mean())
df['online_order'] = df['online_order'].fillna(0)

print("Remaining rows:", len(df))

# MODEL
X = df[['approx_cost(for two people)', 'online_order']]
y = df['rate']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

print("Score:", model.score(X_test, y_test))
print("Prediction:", model.predict([[500, 1]])[0])
```
